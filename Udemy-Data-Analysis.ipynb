# **Data Cleaning**

```python
# check for null values

df.isnull().sum()
```

```python
# check for duplicates

df.duplicated().sum()
```

```python
# rename the price and content duration column to include the units

df.rename(columns={'price': 'price ($)'}, inplace=True)
```

```python
df.rename(columns={'content_duration' : 'content_duration (hrs)'}, inplace=True)
```

```python
# confirm changes

df.info()
```

```python
# drop the course_id as we don't need it for analysis

df.drop('course_id', axis=1, inplace=True)
```

```python
df.info()
```

```python
# convert price to float and timestamp to datetime

df['price ($)'] = df['price ($)'].astype('float64')
```

```python
df['published_timestamp'] = pd.to_datetime(df['published_timestamp'])
```

```python
# confirm changes

df.info()
```

```python
df.head()
```

```python
# export cleaned data

df.to_csv('cleaned_udemy_music_courses_data.csv')
```

```python
plt.bar(df_price_by_courseLevel.index, df_price_by_courseLevel['price ($)'], color='skyblue', edgecolor='black', linewidth=1)
plt.title('Average Price of courses for different course levels')
plt.xlabel('Level')
plt.ylabel('Average Price')
plt.show()
```

```python
df_rating_by_courseLevel = df.groupby(by='level')[['Rating']].mean()
```

```python
df_rating_by_courseLevel = df_rating_by_courseLevel.sort_values(by='Rating', ascending=False)
df_rating_by_courseLevel
```

```python
plt.bar(df_rating_by_courseLevel.index, df_rating_by_courseLevel['Rating'], color='grey', edgecolor='black', linewidth=1)
plt.title('Average Rating of courses for different course levels')
plt.xlabel('Level')
plt.ylabel('Average Rating')
plt.show()
```

```python
df_subscribers_by_courseLevel = df.groupby(by='level')[['num_subscribers']].sum()
```

```python
df_subscribers_by_courseLevel = df_subscribers_by_courseLevel.sort_values(by='num_subscribers', ascending=False)
df_subscribers_by_courseLevel
```

```python
plt.figure(figsize=(20,8))
plt.bar(df_subscribers_by_courseLevel.index, df_subscribers_by_courseLevel['num_subscribers'], color='tan', edgecolor='black', linewidth=1)
plt.title('Total numbers of Subscribers for different course levels')
plt.xlabel('Level')
plt.ylabel('Total Subscribers')
plt.show()
```

`5` How total no of courses published have changed over the years

```python
df.info()
```

```python
# extract year from timeStamp column and create a new column called year published

df['Year Published'] = df['published_timestamp'].dt.year
```

```python
df.head()
```

```python
df_courses_published_perYear = df.groupby(by='Year Published')[['url']].count()
```

```python
df_courses_published_perYear
```

```python
plt.figure(figsize=(20,8))
plt.plot(df_courses_published_perYear.index, df_courses_published_perYear['url'], color='tan', linewidth=2, marker='o', markersize=8, linestyle='--')
plt.title('Trend of courses published per year')
plt.xlabel('Year')
plt.ylabel('Total Number of courses published')
plt.show()
```

# **Summary of Findings**

* The number of courses published steadily grew from 2012 to 2016, there was a decline from 2016 to 2017
